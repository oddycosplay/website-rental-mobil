```python
import re

def map_type(t):
    t_clean = t.strip()
    mapping = {
        'BIGINT UNSIGNED': 'Bigint Unsigned',
        'VARCHAR': 'String',
        'TIMESTAMP': 'Timestamp',
        'ENUM': 'Enum',
        'TEXT': 'Text',
        'BOOLEAN': 'Boolean',
        'DATE': 'Date',
        'YEAR': 'Year',
        'DECIMAL': 'Decimal',
        'INTEGER': 'Integer',
        'DATETIME': 'Datetime',
        'JSON': 'Json'
    }
    return mapping.get(t_clean, t_clean)

def convert_line(line):
    # Match a table row: | attr | TYPE | LENGTH | DESC |
    # (Attr might or might not have backticks)
    match = re.match(r'^\|\s*`?([^`|]+)`?\s*\|\s*([^|]+)\s*\|\s*([^|]+)\s*\|\s*(.*)\s*\|$', line)
    if match:
        attr = match.group(1).strip()
        type_data = map_type(match.group(2).strip())
        length = match.group(3).strip()
        desc = match.group(4).strip()
        # Compact style: exactly one space padding on each side of the content
        return f"| {attr} | {type_data} | {length} | {desc} |"
    return line

def process_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    new_lines = []
    for line in lines:
        stripped = line.rstrip('\n')
        # Check if line is a table divider
        if stripped.strip().startswith('| :') or (stripped.strip().startswith('|') and '---' in stripped):
            # Compact divider style
            new_lines.append("| :--- | :--- | :--- | :--- |")
        else:
            new_lines.append(convert_line(stripped))
            
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write('\n'.join(new_lines) + '\n')

process_file(r'c:\laragon\www\rental_project\Struktur Basis Data.md')
process_file(r'c:\laragon\www\rental_project\documents\DATABASE\Struktur Basis Data.md')
print("Successfully converted files to compact table style!")

```
